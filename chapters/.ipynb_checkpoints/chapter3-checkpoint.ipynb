```python
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pymc3 as pm
import scipy.stats as stats
import seaborn as sns
```

```python
%config InlineBackend.figure_format = 'retina'
az.style.use('arviz-darkgrid')
```

```python
def posterior_grid_approx(grid_points, success, tosses):

    p_grid = np.linspace(0, 1, grid_points)

    prior = np.repeat(5, grid_points)  # uniform
 
    likelihood = stats.binom.pmf(success, tosses, p_grid)

    unstd_posterior = likelihood * prior

    posterior = unstd_posterior / unstd_posterior.sum()
    return p_grid, posterior
```

### 3M5

```python
def posterior_grid_approx(grid_points, success, tosses):

    p_grid = np.linspace(0, 1, grid_points)

    prior = np.zeros(grid_points)
    prior[p_grid >= 0.5] = 5
    prior[p_grid < 0.5] = 0

    likelihood = stats.binom.pmf(success, tosses, p_grid)

    unstd_posterior = likelihood * prior

    posterior = unstd_posterior / unstd_posterior.sum()
    
    return p_grid, posterior
```

```python
p_grid, posterior = posterior_grid_approx(grid_points=10000, success=8, tosses=15)
samples = np.random.choice(p_grid, p=posterior, size=int(1e4), replace=True)
```

```python
plt.plot(p_grid, posterior)
plt.show()
```

```python
az.hdi(samples, 0.9) 
```

```python
resampled = stats.binom.rvs(15, samples, size=10000)
```

```python
az.plot_dist(resampled)
plt.show()
```

```python
len(resampled[resampled == 8])/len(resampled) 
```

```python
new_data = stats.binom.rvs(9, samples, size=10000)
```

```python
len(new_data[new_data == 6])/len(new_data)
```
